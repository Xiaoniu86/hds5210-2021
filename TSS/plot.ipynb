```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
wt=pd.read_csv('./Data/list/combine/ID/combine/all/wt.csv')
dcp=pd.read_csv('./Data/list/combine/ID/combine/all/dcp2.csv')
dhh=pd.read_csv('./Data/list/combine/ID/combine/all/dhh1.csv')
pat=pd.read_csv('./Data/list/combine/ID/combine/all/pat1.csv')
xrn=pd.read_csv('./Data/list/combine/ID/combine/all/xrn1.csv')

wt1=wt.sample(n=200, frac=None, replace=False, weights=None, random_state=None, axis=None)
dcp2=dcp.sample(n=200, frac=None, replace=False, weights=None, random_state=None, axis=None)
dhh1=dhh.sample(n=200, frac=None, replace=False, weights=None, random_state=None, axis=None)
pat1=pat.sample(n=200, frac=None, replace=False, weights=None, random_state=None, axis=None)
xrn1=xrn.sample(n=200, frac=None, replace=False, weights=None, random_state=None, axis=None)
plt.rcParams.update({'figure.figsize':(8,6), 'figure.dpi':100})
```

```python

plt.plot( 'wt1', 'wt2', data=wt1, linestyle='none', marker='o')
plt.plot( 'dhh1_1', 'dhh1_2', data=dhh1, linestyle='none', marker='o')
plt.plot( 'dcp2_1', 'dcp2_2', data=dcp2, linestyle='none', marker='o')
plt.plot( 'pat1_1', 'pat1_2', data=pat1, linestyle='none', marker='o')
plt.plot( 'xrn1_1', 'xrn1_2', data=xrn1, linestyle='none', marker='o')
```
